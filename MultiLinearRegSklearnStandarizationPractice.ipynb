# Create Multiple Linear Regression

# Declare the dependant and independant variable

```python
x = data[['SAT', 'Rand 1,2,3']]
y = data['GPA']
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
x_scaled = scaler.transform(x)
```

```python
x_scaled
```

# Regression with scaled features

```python
reg = LinearRegression()
reg.fit(x_scaled, y)
```

```python
reg.coef_
```

```python
reg.coef_
```

```python
reg.intercept_
```

# Creating a Summaty Table

```python
reg_summary = pd.DataFrame([['Intercept'], ['SAT'], ['Rand 1,2,3']], columns=['Features'])
reg_summary['Weights'] = reg.intercept_, reg.coef_[0], reg.coef_[1]
```

```python
#Weight = Coeficients
reg_summary
```

```python
# ML the word Intercept = Bias
reg_summary = pd.DataFrame([['Bias'], ['SAT'], ['Rand 1,2,3']], columns=['Features'])
reg_summary['Weights'] = reg.intercept_, reg.coef_[0], reg.coef_[1]
```

```python
reg_summary
```

```python
#When we perform feature scalilng, we don't care if a useless variable is there or not
#Either although is better to remove them
```

# Making Predictions with the standardized coefficients (weights)

```python
new_data = pd.DataFrame(data=[[1700,2],[1800,1]],columns=['SAT','Rand 1,2,3'])
new_data
```

```python
#Those values doesn't make sense, there are not a valid GPA
#This happens because the standarization
reg.predict(new_data)
```

```python
#The new data frame should be arranged in the same way
# and also must be standarized in the same way
new_data_scaled = scaler.transform(new_data)
new_data_scaled
```

```python
#this GPA makem more sense
reg.predict(new_data_scaled)
```
