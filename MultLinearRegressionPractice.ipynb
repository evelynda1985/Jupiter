```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
data = pd.read_csv('Files/Sec33/1.02.MultipleLinearRegression.csv')
```

```python
data
```

```python
# GPA = b0 + SAT*b1 + Rand1,2,3*b2
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT','Rand 1,2,3']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

```python
#Simple Linear Regression R2
#R-squared:	0.406
#Adj. R-squared:	0.399

#Multiple Linear regression R2
#R-squared:	0.407
#Adj. R-squared:	0.392

#For having a extra variable the R2 inscrease but not that much
# and the Adj. R2 get lower it means the unsiginificant variable penalized the data
```

```python
#Assumption Number 1
# Linearity: y = b0 + b1x1 + b2x2...bnxn
#the grahphic should look linear
#If the graphic is not linear you should NOT use a linear regression model
#at least you transform the data and after use the linear regression
```

```python
#Assumption Number 2
# No endogeneity 
# Omitted Variable Bias =  When you forgot to include a important variable
# For example for real estate example
# smaller the house higher the prices - Somehting doesn't look right
# where did we draw the sample from? central london
# can we get a better sample? big enough
# why is bigger real estate cheaper? Not in the city of London
# what is it about smaller size that is making it so expensive? --> WE FORGOT LOCATION
# where are the small houses? city of london
# Error term can help to idefy them
```

```python
#Assumption Number 3
# Normality and Homoscedanticity => error ~ N(0, @^2)
# N (Normality) = t	   P>|t| This works because we have assume normality, and the CLT applies
# CLT = central limit theoram applies
# 0 (Zero mean) =  If the error is not expecting be zero so the line is not the best one
# @ (Homoscedascity) = Are we ignoring variables? 
#for example a poor person probably eat everyday egss instead a rich person no jus sometimes
# How to deal with this?
# Look for OVB (Omitted Variables Bias)
# Look for outliers and try to remove them
# log transformation ==> log yhat = b0 + b1x1 (As x increase by 1 unit Y increase by b1 percent)
```

```python
#Assumption Number 4
# No autocorrelation ==> @EiEj = 0: Vi != j
# find patterns if you don't find any you are fine (think on the stocks why mondays people sell and Fridays buy?)
# Detection ==> Durbin-Watson:	0.948 (values between 0 and 4)
# 2 = No autocorrelation
# < 1 and > 3 = cause an alarm
# Don't use linear regression when error terms are correlated
```
